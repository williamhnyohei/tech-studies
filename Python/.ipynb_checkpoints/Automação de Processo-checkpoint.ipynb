## Buscar informações

```python
ticker = input("ticker: ")
```

```python
import yfinance as yf

dados = yf.Ticker(ticker).history("6mo")

fechamento = dados.Close

fechamento.plot()
```

## Criar análise

- analises: cotação máxima, mínima e atual

```python
maxima = round(fechamento.max(), 2)
```

```python
minima = round(fechamento.min(), 2)
```

```python
atual = round(fechamento[-1], 2)
```

```python
maxima
```

```python
minima
```

```python
atual
```

```python
print(f"Resumo das análises: atual: R${atual}; máxima: R${maxima} mínima: R${minima}")
```

```python
import pyautogui
import pyperclip
import time

# abre aba
pyautogui.hotkey("ctrl", "t")

# digitar endereço
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")
time.sleep(2)

pyautogui.click(x=84, y=177)

pyperclip.copy("williamhnyohei@gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

pyperclip.copy("Analises")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

mensagem = f"""
ticker: {ticker}
atual: R${atual};
máxima: R${maxima};
mínima: R${minima}.
"""

pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")
time.sleep(2)

pyautogui.click(x=837, y=686)




```
